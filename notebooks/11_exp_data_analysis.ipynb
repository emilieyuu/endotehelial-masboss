```python
import pandas as pd
import sys
sys.path.append("/Users/emilieyu/endotehelial-masboss")

from src.paths import EXP_DIR, DATA_ANALYSIS_DIR
from src.utils.file_utils import save_df_to_csv, load_csv_to_df

conditions_map = {
    'WT':           ['siCONT_001', 'siCONT_001a', 'siCONT_2'],
    'DSP_KO':       ['siDSP'],
    'TJP1_KO':      ['siTJP1'],
    'JCAD_KO':      ['siJCAD'],
    'DSP_JCAD_DKO': ['siJCAD_siDSP'],
    'TJP1_JCAD_DKO':['siJCAD_siTJP1'],
}
```

## Experiment Key Metrics

```python
"""
Get all key metrics from experiment data (means) for each perturbation.
Combined to dataframe and save. 
"""
rows = []

for condition, files in conditions_map.items():
    dfs = []
    for f in files: 
        try:
            df = load_csv_to_df(EXP_DIR, f) 
            dfs.append(df)
        except Exception as e:
            print(f">>> WARNING: Skipping {f}: {e}")

        
    if not dfs:
        continue

    df = pd.concat(dfs, ignore_index=True)

    ar   = df['Major/Minor Ratio'].dropna()
    area = df['Area'].dropna()
    perimeter = df['Perimeter'].dropna()
    circ = df['Circularity'].dropna()

    row = {
        "perturbation": condition,
        "n": len(ar),
        "mean_ar": ar.mean(),
        "std_ar": ar.std(),
        "mean_area": area.mean(),
        "mean_perimeter": perimeter.mean(),
        "mean_circ": circ.mean(),
    }
    rows.append(row)

combined_df = pd.DataFrame(rows)
save_df_to_csv(combined_df, DATA_ANALYSIS_DIR, "exp_key_metrics")
combined_df
```
