```python
conditions_map = {
    'WT':           ['siCONT_001.csv', 'siCONT_001a.csv', 'siCONT_2.csv'],
    'DSP-KO':       ['siDSP.csv'],
    'TJP1-KO':      ['siTJP1.csv'],
    'JCAD-KO':      ['siJCAD.csv'],
    'DSP-JCAD-DKO': ['siJCAD_siDSP.csv'],
    'TJP1-JCAD-DKO':['siJCAD_siTJP1.csv'],
}
```

```python
print(f"{'Condition':<20} {'n':>5} {'mean_AR':>8} {'std_AR':>7} "
      f"{'mean_area':>10} {'mean_circ':>10}")
print("-" * 65)
exp_targets = {}
for condition, files in conditions_map.items():
    dfs = []
    for f in files:
        path = EXP_DIR / f
        dfs.append(pd.read_csv(path))
        
    if not dfs:
        continue

    df = pd.concat(dfs, ignore_index=True)
    ar   = df['Major/Minor Ratio'].dropna()
    area = df['Area'].dropna()
    circ = df['Circularity'].dropna()

    exp_targets[condition] = {
        'mean_ar':   ar.mean(),
        'std_ar':    ar.std(),
        'median_ar': ar.median(),
        'mean_area': area.mean(),
        'n':         len(ar),
    }
    
    print(f"{condition:<20} {len(ar):>5} {ar.mean():>8.3f} "
          f"{ar.std():>7.3f} {area.mean():>10.1f} {circ.mean():>10.3f}") 
```
